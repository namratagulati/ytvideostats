```python
#importing json file and adding category column
import json
# creating a dictionary that maps id to category
id_to_category = {}
with open('IN_category_id.json', 'r') as f:
    content = json.load(f)
    for c in content['items']:
        id_to_category[c['id']] = c['snippet']['title']
id_to_category
```

```python
df.insert(4, 'category', df['category_id'].map(id_to_category)) #add corresponding category using map fn
df[['category_id', 'category']].head()
```

```python
#find how numerical var are correlated
col = ['views', 'likes', 'dislikes', 'comment_count']
corr_matrix = df[col].corr()
corr_matrix
#close to 1 implies highly correlated (as views increase, likes also increase)
```

```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
fig, ax = plt.subplots()
heatmap = ax.imshow(corr_matrix, interpolation='nearest', cmap='viridis')

# making the colorbar on the side
cbar_min = corr_matrix.min().min()
cbar_max = corr_matrix.max().max()
cbar = fig.colorbar(heatmap, ticks=[cbar_min, cbar_max])
# making the labels
labels = ['']
for i in col:
    labels.append(i)
    labels.append('')
ax.set_yticklabels(labels, minor=False)
ax.set_xticklabels(labels, minor=False)

plt.show()
```

```python
#group by title
df2 = df1.groupby("title").agg({"views":"max","likes":"max","dislikes":"max","comment_count":"max"})
df2.sort_values("views",ascending=False).head(15)
```

Dara Preprocessing
(ensures uniformity of data so that all features contribute equally)

```python
#Scale Variables and Standardization
from sklearn.preprocessing import MinMaxScaler
df2["views_scaled"] = MinMaxScaler(feature_range=(0, 1)).fit(df2[["views"]]).transform(df2[["views"]])
df2["likes_scaled"] = MinMaxScaler(feature_range=(0, 1)).fit(df2[["likes"]]).transform(df2[["likes"]])
df2["dislikes_scaled"] = MinMaxScaler(feature_range=(0, 1)).fit(df2[["dislikes"]]).transform(df2[["dislikes"]])
df2["comment_count_scaled"] = MinMaxScaler(feature_range=(0, 1)).fit(df2[["comment_count"]]).transform(df2[["comment_count"]])
```

```python
# Function of Weighted Sorting Score
# c : comment_count
# v : views
# l : likes
# dl : dislikes

# Weighted Sorting Score : v * c * (l / (l + dl)

def wss(dataframe,c,v,l,dl):
    return (dataframe[v]* dataframe[c]* ((dataframe[l]/ (dataframe[l]+dataframe[dl]))))
```

```python
df2["weighted_sorting_score"] = wss(df2,'comment_count_scaled','views_scaled','likes_scaled','dislikes_scaled')
df2.sort_values("weighted_sorting_score",ascending=False).head(20)
```

The views,likes, dislikes and comment_count columns are no lomger required as we have the scaled columns so these can be deleted

```python
#df2 = df2.drop(["views","likes","dislikes","comment_count"],axis=1)
df2.sort_values("weighted_sorting_score",ascending=False).head(20)
```

```python
#reset the index in another dataset
df3 = df2.reset_index()
df3.sort_values("weighted_sorting_score",ascending=False).head(15)
```

```python
# Category Sorting by Weighted score
df_cat = df.groupby(["category"]).agg({"views":"sum","likes":"sum","dislikes":"sum","comment_count":"sum"})
df_cat.sort_values("views",ascending=False)
```

```python
df_cat["views_scaled"] = MinMaxScaler(feature_range=(0, 1)).fit(df_cat[["views"]]).transform(df_cat[["views"]])
df_cat["likes_scaled"] = MinMaxScaler(feature_range=(0, 1)).fit(df_cat[["likes"]]).transform(df_cat[["likes"]])
df_cat["dislikes_scaled"] = MinMaxScaler(feature_range=(0, 1)).fit(df_cat[["dislikes"]]).transform(df_cat[["dislikes"]])
df_cat["comment_count_scaled"] = MinMaxScaler(feature_range=(0, 1)).fit(df_cat[["comment_count"]]).transform(df_cat[["comment_count"]])
```

```python
df_cat["weighted_sorting_score"] = wss(df_cat,'comment_count_scaled','views_scaled','likes_scaled','dislikes_scaled') #using weighted sorted score function
df_cat = df_cat.drop(["views","likes","dislikes","comment_count"],axis=1)
df_cat.sort_values("weighted_sorting_score",ascending=False)
```

```python
df_cat1 = df_cat.reset_index()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='category',
            y="weighted_sorting_score",
            data=df_cat1,
            order=df_cat1.sort_values('weighted_sorting_score',ascending = False).category.head())

plt.xlabel("Video Category", size=10)
plt.ylabel("Weighted Sorting Score", size=10)
plt.title("Top 5 Video Categories on Youtube", size=15)
plt.tight_layout()
```
